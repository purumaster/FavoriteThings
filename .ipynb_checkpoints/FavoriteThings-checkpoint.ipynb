```python
import pandas as pd
import sqlite3
import os

# create the sqlite DB
db = sqlite3.connect("FavoriteThings.db")

# read CSV into dataframe
df1 = pd.read_csv("FavoriteThings.csv", header="infer")

# convert the pandas dataframe to a sqlite table
df1.to_sql("FavoriteThings", db, if_exists="replace")

def run_query(query):
    return pd.read_sql_query(query,db)


os.getcwd()

```

```python
# view dataframe
# df1
query = ''' 
SELECT * 
FROM FavoriteThings'''
run_query(query)
```

# Favorite Starbucks Drink

```python
query = '''
SELECT DISTINCT FavoriteStarbucksDrink 
FROM   FavoriteThings 
ORDER  BY FavoriteStarbucksDrink ASC;

'''

run_query(query)
```
